```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import kurtosis
#ignore harmless warnings
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.DataFrame(data)
```

```python
df = pd.DataFrame(data)
plt.figure(figsize = (4,4))
labels = ["Delivery Time", "Sorting Time" ]
sns.distplot(df['Delivery Time'], kde = True)
sns.distplot(df['Sorting Time'],hist = True)
plt.legend(labels)
```

```python
kurtosis_values = df.kurtosis()

categories = {}
for col, kurt in kurtosis_values.items():
    if kurt > 3:
        categories[col] = 'Leptokurtic'
    elif kurt < 3:
        categories[col] = 'Platykurtic'
    else:
        categories[col] = 'Mesokurtic'

# Displaying categories and kurtosis values
print("Column \t\t Category \t Kurtosis Value")
print("----------------------------------------------")
for col, category in categories.items():
    print(f"{col} \t {category} \t {kurtosis_values[col]}")
```

```python
# Scatter plot
plt.figure(figsize=(12, 5))
plt.subplot(2, 2, 1)
sns.scatterplot(x='Delivery Time', y='Sorting Time', data=df)
plt.title('Scatter Plot')

```

```python
# Line plot
plt.subplot(2, 2, 2)
sns.lineplot(x=df.index, y='Delivery Time', data=df, marker='o', label='Delivery Time')
sns.lineplot(x=df.index, y='Sorting Time', data=df, marker='o', label='Sorting Time')
plt.title('Line Plot')
plt.legend()

```

```python
# Histogram
plt.subplot(2, 2, 3)
sns.histplot(df['Delivery Time'], bins=3, kde=True, color='skyblue', alpha=0.7)
plt.title('Histogram - Delivery Time')

```

```python
# Boxplot
plt.subplot(2, 2, 4)
sns.boxplot(data=df, orient='h')
plt.title('Boxplot')

plt.tight_layout()
plt.show()          
```
